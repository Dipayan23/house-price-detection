Importin dependencys

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

importing bosten house price dataset

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

Make the dataset into pandas dataframe

```python
house_price_dataframe = pd.DataFrame(house_price_dataset.data,columns=house_price_dataset.feature_names)
house_price_dataframe['MedHouseVal'] = house_price_dataset.target
print(house_price_dataframe.head())
```

checking for missing values

```python
house_price_dataframe.isnull().sum()
```

#Finding some data analytics

```python
house_price_dataframe.describe()
```

Finding the corelation between dataset

```python
correlation = house_price_dataframe.corr()
print(correlation)
```

constructing heatmap

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.2f',annot=True,annot_kws={'size':8},cmap='Blues')
```

Spliting tthe data and target

```python
X = house_price_dataframe.drop(['MedHouseVal'],axis=1)
Y = house_price_dataframe['MedHouseVal']
print(X)
print(Y)
```

Spliting data into test and train

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

Model Training XGBoost Regressor

```python
# Loading the model
model = XGBRegressor()
```

```python
# Train the model with X_train
model.fit(X_train,Y_train)
```

```python
# prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
# Rsqared error
score_1 = metrics.r2_score(Y_train,training_data_prediction)

# mean absolute error
score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)

print(score_1)
print(score_2)
```

```python
# visualize the actual and predict prices
plt.scatter(Y_train,training_data_prediction)
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.title("Actual vs predicted prices")
plt.show()
```

prediction for Test Data

```python
test_data_prediction = model.predict(X_test)
```

```python
# Rsqared error
score_1 = metrics.r2_score(Y_test,test_data_prediction)

# mean absolute error
score_2 = metrics.mean_absolute_error(Y_test,test_data_prediction)

print(score_1)
print(score_2)
```
